```python
from datetime import date


class Persona:
    def __init__(self, nombre = "Sin nombre", apellido = "Sin apellido", fecha_nacimiento = date(1970, 1, 1)):
        self.nombre = nombre
        self.apellido = apellido
        self.fecha_nacimiento = fecha_nacimiento

    def saludos(self):
        print("Saludos, soy", self.nombre, self.apellido)

    def edad(self):
        today = date.today()
        edad = int((today - self.fecha_nacimiento).days / 365)
        return edad

class Estudiante(Persona):
    def __init__(self, nombre = "Sin nombre", apellido = "Sin apellido", fecha_nacimiento = date(1970, 1, 1), grupo = "NA"):
        super().__init__(nombre, apellido, fecha_nacimiento)
        self.grupo = grupo

    def generarMatricula(self):
        return str(self.fecha_nacimiento.year) + self.nombre[:2] + self.apellido[:2]
```

```python
class Registro(Persona):
  def __init__(self, nombre = "Sin nombre", apellido = "Sin apellido", fecha_nacimiento = date(1970, 1, 1)):
    super().__init__(nombre, apellido, fecha_nacimiento)

  def asignar_materias(self, materias=[]): #se asigna tronco común
    tronco_comun=["español", "matematicas", "etica"]
    self.materias = tronco_comun+materias
    return print(f"El estudiante {self.nombre} {self.apellido} tiene asignadas las materias: {self.materias}")

  def registrar_calificacion(self, calificaciones={}):
    registro_cal={}

    for i in self.materias:
      if i in calificaciones:
        registro_cal[i]=calificaciones[i]
      else:
        registro_cal[i]='NA'
    for ii in calificaciones.keys():
      if ii not in self.materias:
        print(f"El estudiante {self.nombre} {self.apellido} no tiene la materia {ii} asignadas, validar asignación primero")

    self.calificaciones=registro_cal

    for m in self.calificaciones.keys():
        print(f"{m}: {self.calificaciones[m]}")

  def asignar_grupo(self, grupo="NA"):
    self.grupo=grupo

  def asignar_semestre(self, semestre="NA"):
    self.semestre=semestre
    return print(f"El estudiante {self.nombre} {self.apellido} ingresó al semestr {self.semestre}")
```

```python
registro1=Registro(nombre ="Francisco",apellido =  "Sanchez", fecha_nacimiento = (1995, 3, 23))
```

```python
registro1.asignar_materias(["evoluación", "antropología"])
```

```python
registro1.registrar_calificacion({'español':8, 'matematicas':9, 'etica':10, 'logica':6})
```
