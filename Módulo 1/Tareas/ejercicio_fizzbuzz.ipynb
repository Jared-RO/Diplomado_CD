```python
##FUNCION DE FIZZBUZZ
def crea_serie(n,m):
  n=int(n)
  m=int(m)
  result=[]
  for i in list(range(n,m+1)):
    if (i%3==0)&(i%5!=0):
      result.append('Fizz')
    elif (i%5==0)&(i%3!=0):
      result.append('Buzz')
    elif (i%3==0)&(i%5==0):
      result.append('Fizz Buzz')
    else:
      result.append(str(i))
  return result

def fizz_buzz():
  x=0
  while x==0:
      print("Declarar n: ")
      n = input()
      print("Declarar m: ")
      m = input()
      print("Usted declaro n =",str(n), " y m =",str(m))
      if (float(n)==int(n)) & (float(m)==int(m)):
          if (int(n)>0) & (int(m)>0):
              if int(n)<int(m):
                x=1
                result=crea_serie(n,m)
                return result
              else:
                print('n debe ser menor a m, repetir')
          else:
              print('n y m deben ser mayores a 0, repetir')
      else:
          print('n y m deben ser enteros, repetir')
```
